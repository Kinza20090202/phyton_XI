```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data_Penjualan_Toko_Online.csv")
print("Data awal:")
print(df.head())
```

```python
df["Diskon"] = 0.10  # 10% dari harga untuk semua produk
```

```python
df["Pendapatan Bersih"] = df["Pendapatan"] - (df["Pendapatan"] * df["Diskon"])
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Produk", y="Pendapatan Bersih", estimator=sum, ci=None)
plt.title("Pendapatan Bersih per Produk (Setelah Diskon 10%)")
plt.xlabel("Produk")
plt.ylabel("Pendapatan Bersih")
plt.show()

```

```python
df.to_csv("analisis_diskon.csv", index=False)
print("✅ Data hasil analisis telah disimpan sebagai 'analisis_diskon.csv'")
```
